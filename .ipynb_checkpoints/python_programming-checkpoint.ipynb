## First method: pass the SQL "interactively"

```python
conn.sql("SELECT * FROM Site LIMIT 5")
```

```python
# Add the fetch to get the data
conn.sql("SELECT * FROM Site LIMIT 5").fetchall() # returns each row as a tuple
```

```python
# Return only one row, good for iterating through table in loops!
conn.sql("SELECT * FROM Site LIMIT 5").fetchone()
```

```python
# Return as a data frame
conn.sql("SELECT * FROM Site LIMIT 5").df()
```

## Using a cursor

```python
# create a cursor
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM Site")
```

```python
# you can applyy the fetchall()..
cur.fetchall()
```

```python
# getting one column from the row
cur.fetchone()[0]
```

```python
# we want to iterate through the rows and print the site code
cur.execute("SELECT * from Site")

for row in cur.fetchall():
    print(row[0])
```

```python
cur.execute("SELECT * from Site")

while True:
    row = cur.fetchone()
    if row == None:
        break
    print(f"Site code: {row[0]}")
```

```python
# you can also create a table
cur.execute("""CREATE TEMP TABLE temp_nests AS
                SELECT * FROM Bird_nests""")
```

```python
cur.execute("SELECT * FROM temp_nests LIMIT 5").fetchall()

# typically execute in one line and THEN fetch call
```
